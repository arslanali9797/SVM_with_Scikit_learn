```python
#ploting laibrery  
import matplotlib.pyplot as plt

import seaborn as sns




```

```python
#we will use the scikit learn laibrery for dataset 
from sklearn.datasets import fetch_lfw_people

# be patince ,it will take time because ,it will download the dataset in first time.
faces = fetch_lfw_people(min_faces_per_person=60,download_if_missing=True,slice_=(slice(70, 195, None),slice(78, 172, None)))

# target names or labels
print(faces.target_names)

# people images 
print(faces.images.shape)

#each image contains 62*47 = 2914 pixles
#y = faces.target
target_names = faces.target_names
n_classes = target_names.shape[0]

print('classe are :',n_classes)



```

```python
# create a grid window which contain 3 rows and 5 col  where images will be shown as a sample along class label
# numpy plot is used for that
fig ,ax =plt.subplots(3,5)
for i ,axi in enumerate(ax.flat):
    axi.imshow(faces.images[i], cmap = 'bone')
    axi.set(xticks=[],yticks=[],
            xlabel
            =faces.target_names[faces.target[i]])
           

 
    
```

```python

# preprocesing to extract jus few important features  and feed that to SVC
from sklearn.svm import SVC

from sklearn.decomposition import PCA as RandomizedPCA

from sklearn.pipeline import make_pipeline

# What is kernal and importance of kernal ,types of kernal
pca =RandomizedPCA(n_components=150, whiten=True,random_state=42)
svc = SVC(kernel='rbf',class_weight='balanced') 
model= make_pipeline(pca,svc)
```

```python
# for the seek of testing the clasifier output


from sklearn.model_selection import train_test_split

# spliting the dataset into train and test form
Xtrain, Xtest , ytrain, ytest = train_test_split(faces.data,faces.target,random_state=42)
```

```python
Xtest[0].shape
```

```python
# we will use grid search cross validation to  explore the combination of parameters
# parameter c is used for control the margin hardness and also  
# another parameter  gima which control the size of our radio basis function (kernal)

from sklearn.model_selection import GridSearchCV

param_grid = {'svc__C':[1,5,10,50],
             'svc__gamma': [0.0001,0.0005,0.001,0.005]}

#now instantiate the model
grid =GridSearchCV(model,param_grid)

#exetution of gridsearch

%time grid.fit(Xtrain,ytrain)

print(grid.best_params_)    
print(grid.best_score_)


# we found the optimal paramertes values  which are 1,0.005

```

```python
model = grid.best_estimator_
yfit =model.predict(Xtest)
```
